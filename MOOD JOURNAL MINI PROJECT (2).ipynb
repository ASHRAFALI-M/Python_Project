# MOOD JOURNAL

```python
import mysql.connector
import tkinter as tk
from tkinter import ttk

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="mood"
)

cursor = conn.cursor()

def log_mood():
    date = date_entry.get()
    mood = mood_entry.get()
    journal_text = journal_text_entry.get("1.0", "end-1c")
    try:
        cursor.execute("INSERT INTO mood_entries (date, mood, journal_text) VALUES (%s, %s, %s)",
                       (date, mood, journal_text))
        conn.commit()
        status_label.config(text="Mood entry logged successfully.")
    except Exception as e:
        status_label.config(text=f"Error logging mood entry: {str(e)}")

def view_mood_entries():
    cursor.execute("SELECT date, mood, journal_text FROM mood_entries ORDER BY date DESC")
    entries = cursor.fetchall()
    if entries:
        result_text.delete("1.0", "end")
        for entry in entries:
            date, mood, journal_text = entry
            result_text.insert("end", f"Date: {date}, Mood: {mood}\n")
            result_text.insert("end", f"Journal Text:\n{journal_text}\n\n")
    else:
        result_text.delete("1.0", "end")
        result_text.insert("end", "No mood entries found.")

def exit_app():
    print("Exiting Mood Journal.")
    conn.close()
    root.destroy()

root = tk.Tk()
root.title("Mood Journal")

frame = ttk.Frame(root)
frame.grid(row=0, column=0, padx=10, pady=10)

date_label = ttk.Label(frame, text="Date (YYYY-MM-DD):")
date_label.grid(row=0, column=0, sticky="w")

date_entry = ttk.Entry(frame)
date_entry.grid(row=0, column=1, padx=5, pady=5)

mood_label = ttk.Label(frame, text="Mood:")
mood_label.grid(row=1, column=0, sticky="w")

mood_entry = ttk.Entry(frame)
mood_entry.grid(row=1, column=1, padx=5, pady=5)

journal_text_label = ttk.Label(frame, text="Journal Text:")
journal_text_label.grid(row=2, column=0, sticky="w")

journal_text_entry = tk.Text(frame, height=5, width=40)
journal_text_entry.grid(row=2, column=1, padx=5, pady=5)

log_button = ttk.Button(frame, text="Log Mood Entry", command=log_mood)
log_button.grid(row=3, column=0, columnspan=2, pady=10)

view_button = ttk.Button(frame, text="View Mood Entries", command=view_mood_entries)
view_button.grid(row=4, column=0, columnspan=2, pady=10)

result_text = tk.Text(frame, height=15, width=40)
result_text.grid(row=5, column=0, columnspan=2, padx=5, pady=5)

status_label = ttk.Label(frame, text="")
status_label.grid(row=6, column=0, columnspan=2)

exit_button = ttk.Button(frame, text="Exit", command=exit_app)
exit_button.grid(row=7, column=0, columnspan=2, pady=10)

if __name__ == "__main__":
    root.mainloop()
```
